```python
from pymongo import MongoClient

# Connexion à la base de données locale MongoDB
client = MongoClient('localhost', 27017)

# Sélection de la base de données 'gym'
db = client['gym']
```

```python
gymnases = db['Gymnases']
for gym in gymnases.find():
    print(gym)
```

```python
sportifs = db['Sportifs']
for gym in sportifs.find():
    print(gym)
```

```python
# a: Trouver les sportifs entre 20 et 30 ans
result_a = db.Sportifs.find(
    {"Age": {"$gte": 20, "$lte": 30}},
    {"_id": 0, "IdSportif": 1, "Nom": 1, "Prenom": 1, "Age": 1}
)
for sportif in result_a:
    print(sportif)

```

```python
# b : Gymnases de "Villetaneuse" ou "Sarcelles" avec une surface > 400 m²
result_b = db.Gymnases.find(
    {"$and": [
        {"$or": [{"Ville": "VILLETANEUSE"}, {"Ville": "SARCELLES"}]},
        {"Surface": {"$gt": 400}}
    ]},
    {"_id": 0, "NomGymnase": 1, "Surface": 1}
)
for gymnase in result_b:
    print(gymnase)
```

```python
# c: Trouver les sportifs qui pratiquent du hand ball
result_c = db.Sportifs.find(
    {"Sports.Jouer": "Hand ball"},
    {"_id": 0, "IdSportif": 1, "Nom": 1}
)
for sportif in result_c:
    print(sportif)

```

```python
# d: Trouver les gymnases et les jours où il y a des séances de hand ball
result_d = db.Gymnases.find(
    {"Seances.Libelle": "Hand ball"},
    {"_id": 0, "NomGymnase": 1, "Seances.$": 1}
)
for gymnase in result_d:
    print(gymnase)
```

```python
# e: Trouver les gymnases où l'on peut jouer au hockey le mercredi après 15h
result_e = db.Gymnases.find(
    {"$and": [
        {"Seances.Libelle": "Hockey"},
        {"Seances.Jour": "mercredi"},
        {"Seances.Horaire": {"$gt": 15}}
    ]},
    {"_id": 0, "NomGymnase": 1, "Seances.$": 1}
)
for gymnase in result_e:
    print(gymnase)
```

```python
# f: Trouver les sportifs qui ne pratiquent aucun sport
result_f = db.Sportifs.find(
    {"Sports": {"$exists": False}},
    {"_id": 0, "IdSportif": 1, "Nom": 1}
)
for sportif in result_f:
    print(sportif)
```

```python
# g: Trouver les gymnases qui n'ont pas de séances le dimanche
result_g = db.Gymnases.find(
    {"Seances.Jour": {"$ne": "dimanche"}},
    {"_id": 0, "NomGymnase": 1}
)
for gymnase in result_g:
    print(gymnase)
```

```python
# h: Trouver les gymnases qui ne proposent que des séances de basket ball ou de volley ball
result_h = db.Gymnases.find(
    {"Seances.Libelle": {"$all": ["Basket ball", "Volley ball"]}},
    {"_id": 0, "NomGymnase": 1}
)
for gymnase in result_h:
    print(gymnase)
```

```python
# h: Trouver les gymnases qui ne proposent que des séances de basket ball ou de volley ball
result_h = db.Gymnases.find(
    {"Seances": {
        "$not": {
            "$elemMatch": {
                "Libelle": {"$nin": ["Basket ball", "Volley ball"]}
            }
        }
    }},
    {"_id": 0, "NomGymnase": 1, "Seances.Libelle": 1}
)
for gymnase in result_h:
    print(gymnase)
```

```python
# i : Sportifs qui sont à la fois joueurs et entraîneurs
result_i = db.Sportifs.find(
    {"$and": [
        {"Sports.Jouer": {"$exists": True}},
        {"Sports.Entrainer": {"$exists": True}}
    ]},
    {"_id": 0, "IdSportif": 1, "Nom": 1, "Prenom": 1, "Sports": 1}
)
for sportif in result_i:
    print(sportif)
```

```python
# j : Sportifs qui sont des conseillers
result_j = db.Sportifs.find(
    {"IdSportifConseiller": {"$exists": True}},
    {"_id": 0, "IdSportif": 1, "Nom": 1, "Prenom": 1}
)
for sportif in result_j:
    print(sportif)
```

```python
# k : Trouver le conseiller de "Kervadec"
kervadec = db.Sportifs.find_one({"Nom": "KERVADEC"}, {"IdSportifConseiller": 1})

if kervadec:
    conseiller = db.Sportifs.find_one({"IdSportif": kervadec["IdSportifConseiller"]}, {"Nom": 1, "Prenom": 1})
    print(conseiller)
else:
    print("Kervadec non trouvé.")
```

```python
# l : Entraîneurs qui entraînent à la fois le hand ball et le basket ball
result_l = db.Sportifs.find(
    {"$and": [
        {"Sports.Entrainer": {"$all": ["Hand ball", "Basket ball"]}}
    ]},
    {"_id": 0, "IdSportif": 1, "Nom": 1, "Prenom": 1}
)
for entraineur in result_l:
    print(entraineur)
```

```python
# m: Trouver les couples de sportifs de même âge

# Récupérer tous les sportifs avec seulement les champs nécessaires
sportifs = list(db.Sportifs.find({}, {"IdSportif": 1, "Nom": 1, "Prenom": 1, "Age": 1}))

# Grouper les sportifs par âge
sportifs_par_age = {}
for sportif in sportifs:
    sportifs_par_age.setdefault(sportif['Age'], []).append(sportif)

# couples avec le même âge
couples = []
for sportifs in sportifs_par_age.values():
    if len(sportifs) > 1:  
        for i in range(len(sportifs)):
            for j in range(i + 1, len(sportifs)):
                s1 = sportifs[i]
                s2 = sportifs[j]
                couples.append((s1['IdSportif'], s1['Nom'], s1['Prenom'], s2['IdSportif'], s2['Nom'], s2['Prenom']))

# Afficher les couples
for couple in couples:
    print(f"Sportif 1: ID={couple[0]}, Nom={couple[1]}, Prénom={couple[2]} | "
          f"Sportif 2: ID={couple[3]}, Nom={couple[4]}, Prénom={couple[5]}")
```

```python
# n : Moyenne d'âge des sportives qui pratiquent du basket ball
result_n = db.Sportifs.aggregate([
    {"$match": {"Sports.Jouer": "Basket ball", "Sexe": "F"}},
    {"$group": {"_id": None, "moyenne_age": {"$avg": "$Age"}}}
])
for res in result_n:
    print(res)
```

```python
# o: Trouver le sportif le plus jeune
result_o = db.Sportifs.find().sort("Age", 1).limit(3)
for sportif in result_o:
    print(sportif)
```

```python
# p : Gymnase de "Stains" ou "Montmorency" avec la plus grande surface
result_p = db.Gymnases.find(
    {"Ville": {"$in": ["STAINS", "MONTMORENCY"]}},
    {"_id": 0, "NomGymnase": 1, "Surface": 1}).sort("Surface", -1).limit(3)
for gymnase in result_p:
    print(gymnase)

```

```python
# q : Entraîneurs qui n'entraînent que du hand ball ou du basket ball
result_q = db.Sportifs.find(
    {"$and": [
        {"Sports.Entrainer": {"$all": ["Hand ball", "Basket ball"]}},
        {"Sports.Entrainer": {"$size": 2}}  
    ]},
    {"_id": 0, "IdSportif": 1, "Nom": 1, "NomGymnase": 1, "Sports": 1}
)
for entraineur in result_q:
    print(entraineur)

```

```python
# r: Trouver les couples de sportifs de même âge avec le même conseiller

result_r = db.Sportifs.aggregate([
    # Grouper par Age et IdSportifConseiller
    {"$group": {
        "_id": {"Age": "$Age", "IdSportifConseiller": "$IdSportifConseiller"},
        "sportifs": {"$push": {"IdSportif": "$IdSportif", "Nom": "$Nom", "Prenom": "$Prenom"}},
        "count": {"$sum": 1}
    }},
    {"$match": {"count": 2}}
])

for group in result_r:
    age = group['_id']['Age']
    conseiller = group['_id']['IdSportifConseiller']
    sportifs = group['sportifs']
    s1 = sportifs[0]
    s2 = sportifs[1]
    
    print(f"Couple avec âge {age} et conseiller {conseiller}: "
          f"Sportif 1: ID={s1['IdSportif']}, Nom={s1['Nom']}, Prénom={s1['Prenom']} | "
          f"Sportif 2: ID={s2['IdSportif']}, Nom={s2['Nom']}, Prénom={s2['Prenom']}")
```

```python
# s: Trouver les sportifs sans conseillers
result_s = db.Sportifs.find(
    {"IdSportifConseiller": {"$exists": False}},
    {"_id": 0, "IdSportif": 1, "Nom": 1}
)
for sportif in result_s:
    print(sportif)
```

```python
# t : Nombre de sports qu'arbitre chaque sportif
result_t = db.Sportifs.aggregate([
    {
        "$unwind": "$Sports.Arbitrer"
    },
    {
        "$group": {
            "_id": "$IdSportif", 
            "count": {"$sum": 1}     
        }
    }
])

# Affichage des résultats
for sportif in result_t:
    print(f"Sportif ID={sportif['_id']}, Nombre de sports arbitrés: {sportif['count']}")
```

```python
# u : Horaires des premières et dernières séances pour chaque gymnase de Stains
result_u = db.Gymnases.aggregate([
    {"$match": {"Ville": "STAINS"}},
    {"$unwind": "$Seances"},
    {"$group": {
        "_id": "$NomGymnase",
        "premiere_heure": {"$min": "$Seances.Horaire"},
        "derniere_heure": {"$max": "$Seances.Horaire"}
    }}
])
for gymnase in result_u:
    print(gymnase)
```

```python
# v : Nombre de séances journalières assurées par chaque entraîneur de hand ball
entraineurs = list(db.Sportifs.find({"Sports.Entrainer": "Hand ball"}, {"IdSportif": 1, "Nom": 1, "Prenom": 1}))

# compter les séances de Hand ball qu'il dirige
for entraineur in entraineurs:
    seances = list(db.Gymnases.find({
        "Seances.IdSportifEntraineur": entraineur['IdSportif'],
        "Seances.Libelle": "Hand ball"
    }, {"Seances": 1}))

    # Compter les séances par jour
    seances_par_jour = {}
    for gymnase in seances:
        for seance in gymnase["Seances"]:
            if seance["IdSportifEntraineur"] == entraineur['IdSportif'] and seance["Libelle"] == "Hand ball":
                jour = seance["Jour"]
                seances_par_jour[jour] = seances_par_jour.get(jour, 0) + 1

    # Affichage du résultat pour chaque entraîneur
    print(f"Entraîneur: {entraineur['Prenom']} {entraineur['Nom']}")
    for jour, nombre in seances_par_jour.items():
        print(f"  {jour}: {nombre} séance(s)")
    print()
```

```python
# w. Requête pour trouver les gymnases ayant plus de 15 séances le mercredi
result_w = db.Gymnases.aggregate([
    {"$unwind": "$Seances"},
    {"$match": {"Seances.Jour": "mercredi"}},
    {"$group": {
        "_id": "$NomGymnase",
        "nombre_seances": {"$sum": 1}
    }},
    {"$match": {"nombre_seances": {"$gt": 15}}}
])
for gymnase in result_w:
    print(gymnase)
```

```python
# w. Requête pour trouver les gymnases ayant plus de 15 séances le mercredi
result_w = db.Gymnases.aggregate([
    {"$unwind": "$Seances"}, 
    {"$match": {
    #"$regex": "^mercredi$" =  sans autres caractères avant ou après ,"$options": "i" =  insensible à la casse(maj,minuscule)
        "Seances.Jour": {"$regex": "^mercredi$", "$options": "i"}
    }},
    {"$group": {
        "_id": "$IdGymnase",  
        "nombreSeancesMercredi": {"$sum": 1} 
        }}])

for gymnase in result_w:
    print(gymnase)
```
